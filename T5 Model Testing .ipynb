```python
import pandas as pd
from pipelines import pipeline
excel_file = pd.read_excel(r'C:\Users\Anura\OneDrive\Amaris.AI Product & Project Mgmt Internship\Testing\json_to_MCQ_99 privacy.xlsx',sheet_name='Sheet1')
```

```python
pd.DataFrame(excel_file)
```

```python
df = pd.DataFrame(excel_file)
df_nonull = df[~df['text'].isna()] #removes empty rows in excel file input 

nlp1 = pipeline("multitask-qa-qg", model="valhalla/t5-base-qa-qg-hl") #calling in multitask qa-qg pipeline 

for index, row in df_nonull.iterrows(): #iterates through all rows in text column 
    try:
        payload = row['text']
        counter = 1
        for i in [nlp1]:
            output = str(i(payload))
            col_name = 'T5 finetuned model (unclean text): Multitask QA_QG Base Model'
            df.loc[index, col_name] = str(output)
            counter +=1
    except ValueError:
        print(i)
        #df_nonull.loc[i, 'questgen.ai_bool'] = 'error'

#for i in range(14):
    #nlp1 = df['text'][i]
    #output = nlp(nlp1) 
    #df.loc[i, 'T5 finetuned model (unclean text): MultiTask QA-QG'] = str(output) #need to change 
    #print(output)
```
